```python
#Import dependencies
import pandas as pd
```

```python
#Read in cleaned CSV as DataFrame
delhi_df = pd.read_csv('delhi_weather_cleaned.csv')
delhi_df
```

```python
#Creating the datetime/conds DataFrame
conds_df = delhi_df[['datetime_utc', 'conds']].set_index('datetime_utc')
conds_df
```

```python
#Creating the features DataFrame
features_df = delhi_df[['datetime_utc', 'dewptm', 'fog', 'hail', 'hum', 'pressurem',
       'rain', 'snow', 'tempm', 'thunder', 'tornado', 'vism', 'wdird', 'wdire',
       'wspdm']].set_index('datetime_utc')
features_df
```

```python
features_df.columns
```

```python
#Exporting DataFrames as .csv files for exporting to delhi.db
conds_df.to_csv('Database/conds.csv',mode='w', header=False)
features_df.to_csv('Database/features.csv',mode='w', header=False)
```
