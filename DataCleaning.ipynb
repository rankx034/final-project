```python
#Import dependencies
import pandas as pd
```

```python
#Read the CSV as a DataFrame 
delhi_df = pd.read_csv("delhi_weather.csv")

delhi_df
```

```python
#Display data types
delhi_df.dtypes
```

```python
#Removing all leading space and underscores in column names
delhi_df.columns = delhi_df.columns.str.replace(' _', '')
delhi_df
```

```python
#Counting NaN values to determine which columns can be dropped
delhi_df.isna().sum()
```

```python
#Dropping columns that cannot or will not be used
delhi_cleaned = delhi_df.drop(['heatindexm','precipm','wgustm','windchillm'], axis=1)
delhi_cleaned.isna().sum()
```

```python
#Dropping all rows with NaN now that the columns with mostly NaN 
#will not cause us to drop almost every row
delhi_cleaned = delhi_cleaned.dropna()
```

```python
#Checking again for NaN
delhi_cleaned.isna().sum()
```

```python
#Determining all conds to then categorize them
delhi_cleaned['conds'].value_counts()
```

```python
#Dropping rows with "Unknown" conditions
delhi_cleaned = delhi_cleaned[delhi_cleaned.conds != 'Unknown']
```

```python
#Checking that 'Unknown' condition was removed
delhi_cleaned['conds'].value_counts()
```

```python
columns_to_haze = ['Fog', 'Shallow Fog', 'Smoke', 'Heavy Fog', 'Blowing Sand', 'Light Fog', 'Light Haze', 'Patches of Fog', 'Sandstorm', 'Light Sandstorm', 'Partial Fog', 'Volcanic Ash', 'Widespread Dust']    
columns_to_rain = ['Drizzle', 'Partial Rain', 'Light Rain', 'Heavy Rain', 'Heavy Thunderstorms and Rain', 'Heavy Thunderstorms with Hail', 'Light Drizzle', 'Light Rain Showers', 'Light Thunderstorm', 'Light Thunderstorms and Rain', 'Rain Showers', 'Thunderstorm', 'Thunderstorms and Rain', 'Thunderstorms with Hail', 'Light Hail Showers', 'Mist', 'Light Freezing Rain', 'Squalls'] 
columns_to_cloudy = ['Overcast', 'Mostly Cloudy', 'Partly Cloudy', 'Scattered Clouds', 'Funnel Cloud']
```

```python
columns_to_rain
```

```python
#Consolidating conditions
delhi_categories = delhi_cleaned.replace(columns_to_haze, 'Haze')
delhi_categories = delhi_categories.replace(columns_to_rain, 'Rain')
delhi_categories = delhi_categories.replace(columns_to_cloudy, 'Cloudy')                                         
#delhi_cleaned['conds'] = delhi_cleaned.conds.map({columns_to_haze:'Haze'})
```

```python
delhi_categories['conds'].value_counts()
```

```python
#Exporting the cleaned DataFrame as a CSV to serve as the provisional database
delhi_categories.to_csv("delhi_weather_cleaned.csv")
```
